## Desafio 01: Investigar o motivo do desbalanceamento da classe "tratamento"
Para verificar a questão do desbalanceamento nos dados de tratamento, é preciso explorar melhor os dados de tratamento e drogas,  para ver se existe alguma relação. Vamos explorar um pouco mais:

##### 1º Fato Relevante
Analisando as 4 linhas a seguir, fica evidente que o tratamento "com_controle" utilizou somente a droga `cacb2b860` e os dados de dose e tempo estão balanceados. 

```python
# Isolar os dados dos tratamentos "com_controle"
dados_cctrl = dados[dados.tratamento == "com_controle"] # Filtrar os dados de tratamento "com_controle" e armazenar na variável "dados_cctrl"
dados_cctrl.tratamento.value_counts() # Contar a quantidade de valores no tratamento "com_controle"
```

```python
# Verificar quais os valores exclusivos da coluna "droga"
dados_cctrl.droga.unique()
```

```python
dados_cctrl['dose'].value_counts(normalize = True)  # Verificando a proporção de doses no tratamento "com_controle"
```

```python
dados_cctrl['tempo'].value_counts(normalize = True) # Verificando a proporção do tempo no tratamento "com_controle"
```

###2º Fato Relevante
Nas 7 linhas a seguir, fica evidenciado que a droga `cacb2b860` não foi utilizada para o tratamento "com_droga" e os dados de dose e tempo estão balanceados. 

```python
# Isolar os dados dos tratamentos "com_droga"
dados_cdrg = dados[dados.tratamento == "com_droga"] # Filtrar os dados de tratamento "com_droga" e armazenar em uma variável
dados_cdrg.tratamento.value_counts() # Contar a quantidade de valores no tratamento "com_droga"
```

```python
# Verificar quais os valores exclusivos da coluna "droga"
dados_cdrg.droga.unique()
```

```python
dados_cdrg['dose'].value_counts(normalize = True) # Verificando a proporção de doses no tratamento "com_drogas"
```

```python
dados_cdrg['tempo'].value_counts(normalize = True) # Verificando a proporção do tempo no tratamento "com_drogas"
```

```python
# Verificar a quantidade de valores para cada classe da coluna "droga"
dados_cdrg.droga.value_counts()
```

```python
dados_cdrg[dados_cdrg.droga == 'cacb2b860'].value_counts() # Não existem experimentos realizados com a droga "cacb2b860"
```

```python
# Visualizar a Distribuição das Classes da Coluna tratamento com todos os dados
total_tratamento = dados['tratamento'].value_counts()
print('Classe 1:', total_tratamento[0])
print('Classe 2:', total_tratamento[1])
print('Proporção da Classe 0 para Classe 1:', round(total_tratamento[0] / total_tratamento[1], 2), ': 1')
total_tratamento.plot(
    kind='bar', 
    title='Distribuição Classes Tratamento', 
    xlabel='Classes Tratamento', 
    ylabel='Qtd. de linhas',
    color = ['#4d1d4d', '#D7410E']);
```

## Resultado
Com isso, entendo que os dados de tratamento "com_controle" não estão balanceados pelo fato de simplesmente representararem o grupo de controle do experimento, o qual fornece uma referência utilizando apenas a droga `cacb2b860`. Fato é que os dados de dose e tempo estão com balanceamento dentro do aceitável para o experimento em questão.

## Desafio 07: Personalizar os gráficos por meio do Matplotlib
Aqui o objetivo é deixar os gráficos mais atraentes por meio da biblioteca Matplotlib. Documentação disponível em: https://matplotlib.org/stable/contents.html

```python
# Desafio 07 - Gráfico 01 - Classes Tratamento
dados_tratamento = dados['tratamento'].value_counts()
dados_tratamento.plot.pie(
    figsize=(8, 8), # Definição do tamanho do gráfico em polegadas
    labels=["Com droga","Com controle"],  # Definição do títulos dos eixos X e Y
    title='Distribuição Classes Tratamento', # Definição do título do gráfico
    colors=['#4d1d4d', '#D7410E'], # Definição das cores de cada fatia
    fontsize=14, # Definição do tamanho da fonte dos valores do eixo X e Y
    explode=(0,0.2), # Dá um destaque na segunda fatia do gráfico
    autopct="%.1f"); # Exibe o percentual no forma da string passada
```

```python
# Desafio 07 - Gráfico 02 - Classes Tempo
dados_tempo = dados['tempo'].value_counts()
dados_tempo.plot.bar(
    title='Distribuição Classes Tempo', # Definição do título do gráfico
    xlabel='Classes Tempo', # Definição do título do eixo X
    ylabel='Qtd Total de Linhas', # Definição do título do eixo Y
    color = ['#4d1d4d', '#054F77','#D7410E'], # Definição das cores de cada barra
    figsize=(8,4), # Definição do tamanho do gráfico em polegadas
    fontsize=12); # Definição do tamanho da fonte dos valores do eixo X e Y
```
